```python
import pandas as pd

def analizar_con_diccionario(ruta_csv, ruta_excel):
    print("==========================================================")
    print("🚀 Iniciando Análisis de Datos con tu Diccionario de Excel")
    print("==========================================================\n")
    
    # 1. Cargamos tu diccionario desde el archivo de Excel (.xlsx)
    try:
        df_diccionario = pd.read_excel(ruta_excel)
        # Limpiamos espacios en blanco externos en los nombres de las columnas
        df_diccionario.columns = df_diccionario.columns.str.strip()
    except Exception as e:
        print(f"❌ Error al intentar abrir el archivo de Excel: {e}")
        return

    # --- DETECTOR INTELIGENTE DE COLUMNAS ---
    # Buscamos la columna que contenga "Nombre de la columna" y "Descripción" sin importar el texto largo
    col_llave = None
    col_valor = None

    for col in df_diccionario.columns:
        if "Nombre de la columna" in col:
            col_llave = col
        if "Descripción" in col or "Descripcion" in col:
            col_valor = col

    # Si por alguna razón el detector automático no las encuentra, usamos los nombres completos:
    if not col_llave:
        col_llave = 'Nombre de la columna\n\n(Nombre de la columna tal y como aparece en la base)'
    if not col_valor:
        col_valor = 'Descripción\n\n(Breve descripción de qué información contiene la columna)'
    # ----------------------------------------

    # Creamos el mapeo seguro relacionando el nombre de la columna con su descripción
    diccionario_mapeo = dict(zip(df_diccionario[col_llave], df_diccionario[col_valor]))

    # 2. Cargamos tu archivo principal de datos (.csv)
    try:
        df_datos = pd.read_csv(ruta_csv)
    except Exception as e:
        print(f"❌ Error al intentar abrir el archivo CSV: {e}")
        return

    print(f"📊 Dataset cargado con éxito. Filas: {df_datos.shape[0]} | Columnas: {df_datos.shape[1]}\n")

    # 3. Recorremos cada columna del CSV y hacemos el cruce con el diccionario
    print("----------------------------------------------------------")
    print("🔍 ANÁLISIS INDIVIDUAL POR COLUMNA:")
    print("----------------------------------------------------------")
    
    for columna in df_datos.columns:
        print(f"\n🔹 Columna: {columna.upper()}")
        
        # Buscamos qué significa esta columna según nuestro Excel
        descripcion = diccionario_mapeo.get(columna, "⚠️ No se encontró descripción en el archivo de Excel.")
        print(f"   📝 Descripción del diccionario: {descripcion}")
        
        # Extraemos métricas y datos de calidad de la columna
        tipo_dato = df_datos[columna].dtype
        nulos = df_datos[columna].isnull().sum()
        porcentaje_nulos = (nulos / len(df_datos)) * 100
        
        print(f"   ⚙️ Tipo de dato técnico: {tipo_dato}")
        print(f"   🔲 Valores faltantes (nulos): {nulos} ({porcentaje_nulos:.2f}%)")
        
        # Si la columna es numérica (latitud, longitud, año, etc.)
        if pd.api.types.is_numeric_dtype(df_datos[columna]):
            print(f"   📈 Estadísticas rápidas:")
            print(f"      - Valor mínimo: {df_datos[columna].min()}")
            print(f"      - Valor máximo: {df_datos[columna].max()}")
            print(f"      - Promedio general: {df_datos[columna].mean():.4f}")
            
        # Si la columna es de texto o categorías (alcaldía, folio, tipo de incidente)
        else:
            valores_unicos = df_datos[columna].nunique()
            print(f"   🔤 Análisis de categorías:")
            print(f"      - Total de categorías únicas: {valores_unicos}")
            if valores_unicos > 0:
                print(f"      - Top 3 valores más recurrentes:")
                top_3 = df_datos[columna].value_counts().head(3)
                for valor, conteo in top_3.items():
                    print(f"        • '{valor}': {conteo} veces")

if __name__ == "__main__":
    # 📌 Configura aquí los nombres exactos de tus archivos locales:
    mi_archivo_csv = "viales_limpio.csv"
    mi_diccionario_excel = r"C:\Users\julio\Documents\proyectos\Proyecto APADD\Proyecto\datos\diccionario-incidentes-viales-c5 (1).xlsx"
    
    analizar_con_diccionario(mi_archivo_csv, mi_diccionario_excel)
```
